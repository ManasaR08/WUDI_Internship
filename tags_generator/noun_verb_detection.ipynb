```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
f = open("preprocessed_text.txt", "r",encoding="utf-8")
text = f.read()
```

```python
text_n = nlp(text)
```

```python
nouns=[]
for chunk in text_n.noun_chunks:
    nouns.append(chunk)
```

```python
import re

mystr = text
words = re.sub("[^\w]", " ",  mystr).split()
```

```python
words
```

```python
from nltk.corpus import wordnet as wn
pos_all = dict()
for w in words:
    pos_l = set()
    for tmp in wn.synsets(w):
        if tmp.name().split('.')[0] == w:
            pos_l.add(tmp.pos())
    pos_all[w] = pos_l
print (pos_all)
```

```python
import nltk
from nltk.tokenize import word_tokenize

text = word_tokenize(text)
```
